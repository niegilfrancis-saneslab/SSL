```python
import glob
import re
import os
import tqdm
```

```python
experiment_no = 384
folder = f"D:/big_setup/experiment_{experiment_no}/concatenated_data_cam_mic_sync/"
```

```python
# Functions to sort file names correctly
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [atoi(c) for c in re.split(r'(\d+)', text)]
```

```python
try:
    os.makedirs(folder+"tracks/")
except:
    pass
```

```python
# get all the slp files
slp_files = glob.glob(folder+"*pred_neck_anchor.slp")
slp_files.sort(key=natural_keys)
```
